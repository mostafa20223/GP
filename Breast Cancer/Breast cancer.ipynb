## The indexes of proteins 
##########4965 -----------> index:11

```python
Protein_images = b_data['MSI_data_cube']
```

```python
pro_indexes = [11,12,13]
sp_mz = [4965,4999,5067]
```

```python
plt.figure(figsize = (60,60))
for i in range(0,3):
    sp = plt.subplot(3,1,i+1)
    sp.axis('off')
    sp.set_title('m/z = '+str(sp_mz[i]),fontsize=100)
    plt.imshow(Protein_images[:,:,pro_indexes[i]],cmap='jet')
    plt.colorbar()
plt.show()
    
```
